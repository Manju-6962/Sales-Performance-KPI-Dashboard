```python
import pandas as pd
from google.colab import files
import os

file_name = "Sales.csv.xlsx"

# Check if the file exists. If not, prompt the user to upload it.
if not os.path.exists(file_name):
    print(f"The file '{file_name}' was not found. Please upload it.")
    uploaded = files.upload()
    # Check if the desired file was actually uploaded
    if file_name not in uploaded:
        print(f"Warning: '{file_name}' was not found among the uploaded files. Please re-run this cell and ensure you upload the correct file.")
    else:
        print(f"'{file_name}' uploaded successfully.")
        # Proceed with reading the file now that it's uploaded
        df = pd.read_excel(file_name)
        print("Successfully read the Excel file. Here are the first 5 rows:")
        print(df.head())
else:
    print(f"The file '{file_name}' already exists. Reading the file.")
    df = pd.read_excel(file_name)
    print("Successfully read the Excel file. Here are the first 5 rows:")
    print(df.head())
```
